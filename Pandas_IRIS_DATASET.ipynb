```python
#pehle nan values kuch place apne data mei
import numpy as np
df=iris.copy()
df.iloc[2:4,1:3]=np.nan
df.head(10)
```

```python
df.describe()
```

```python
#so 148 in sl and sw as 2 nan values
#now drop the rows with nan values
df.dropna(inplace=True)
df.head()
```

```python
#so row 2 and 3 removed
#now again nan place
df.iloc[2:4,1:3]=np.nan
#aur see how to fill nan with mean
df.SepalLengthCm.fillna(df.SepalLengthCm.mean(),inplace=True) #so sl column ke nan ko ussi column ke mean se fill
df.head()
```

```python
#now want ki sw as both nan of species setosa so only setosa wallo ki mean se fill
setosa_data=df[df.Species=='Iris-setosa']
df.SepalWidthCm.fillna(setosa_data.SepalWidthCm.mean(),inplace=True)
df.head(10)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.head(10)
```

```python
del df['Id']
df.head(10)
```
